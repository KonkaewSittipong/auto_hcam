```python
import os
import glob
import hipercam.scripts as scripts

# ANSI Colors for terminal output
BLUE = '\033[94m'
RED = '\033[91m'
RESET = '\033[0m'

class hipercam_setup:
    def __init__(self, source, save_dir, raw_dir, input_run_ul=None, ccd='1'):
        # Calibration tracking (Initialized as empty lists for .extend)
        self.bias = {'lis': [], 'master': None}
        self.dark = {'lis': [], 'master': None}
        self.flat = {'lis': [], 'master': None}
        self.data = {'lis': []}
        
        self.raw_dir = raw_dir
        self.base_dir = save_dir
        self.source = source
        self.ccd = ccd
        
        if self.base_dir:
            os.makedirs(self.base_dir, exist_ok=True)
        
        # Branching logic for different sources
        if source == 'ul':
            if input_run_ul:
                print(f'{BLUE}Working with source: ULTRACAM Local (ul){RESET}')
                self.input_run_ul = input_run_ul
                self.source_ul()
            else:
                print(f'{RED}Error: Source "ul" requires input_run_ul dictionary.{RESET}')

        elif source == 'hf':
            print(f'{BLUE}Working with source: HiPERCAM Local (hf){RESET}')
            # Logic: Just glob existing .hcm files
            self.bias['lis'] = glob.glob(os.path.join(self.raw_dir, "bias*.hcm"))
            
        elif source == 'us':
            print(f'{BLUE}Working with source: ULTRACAM Server (us){RESET}')
            # Logic: Similar to 'ul' but run_path is just the run ID
            self.source_ul() 

    def grab(self, work_dir, run_path, prefix_out, **kwargs):
        """Wraps hipercam.scripts.grab and tracks resulting .hcm files."""
        if not os.path.exists(work_dir):
            os.makedirs(work_dir, exist_ok=True)
            
        os.chdir(work_dir)
        run_name = os.path.basename(run_path)
        
        args = [
            "nodef", f"source={self.source}", f"run={run_path}", 
            f"first={kwargs.get('f1', 1)}", f"last={kwargs.get('f2', 0)}", 
            f"ndigit={kwargs.get('ndigit', 3)}", f"trim={kwargs.get('trim', False)}", 
            f"bias={kwargs.get('bias', 'none')}", f"dark={kwargs.get('dark', 'none')}",
            f"flat={kwargs.get('flat', 'none')}", "fmap=none",
            f"output={prefix_out}{run_name}"
        ]
        
        try:
            print(f"{BLUE}#### Grabbing {run_name} ####{RESET}")
            scripts.grab(args)
            
            # Find the converted files (e.g., bias_all_run002_001.hcm)
            found_files = glob.glob(f"{prefix_out}{run_name}_*.hcm")
            
            # Record absolute paths into tracking lists
            full_paths = [os.path.join(work_dir, f) for f in found_files]
            if 'bias' in prefix_out:
                self.bias['lis'].extend(full_paths)
            elif 'dark' in prefix_out:
                self.dark['lis'].extend(full_paths)
            elif 'flat' in prefix_out:
                self.flat['lis'].extend(full_paths)
                
        except Exception as e:
            print(f"{RED}Error on {run_name}: {e}{RESET}")

    def source_ul(self):
        """Processes the nested dictionary structure (bias/dark/flat -> subkey)"""
        # Process categories in order
        for cat in ['bias', 'dark', 'flat', 'data']:
            if cat not in self.input_run_ul: continue
            
            cat_dir = os.path.join(self.base_dir, cat)
            
            # Iterate through sub-configurations (e.g., 'all', 'science_lis')
            for sub_key, config in self.input_run_ul[cat].items():
                if config is None: continue
                
                # Check for specific raw_dir in config, else use default
                rawdir = config.get('raw_dir', self.raw_dir)
                runs = config.get('runs', [])
                frames = config.get('frames', [])
                
                for i, run in enumerate(runs):
                    # Handle frame range [f1, f2]
                    f_range = frames[i] if isinstance(frames[0], list) else frames
                    
                    self.grab(
                        work_dir=cat_dir,
                        run_path=os.path.join(rawdir, run) if self.source == 'ul' else run,
                        prefix_out=f"{cat}_{sub_key}_",
                        f1=f_range[0],
                        f2=f_range[1]
                    )
```
